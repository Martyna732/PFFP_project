```python
import numpy as np
from main import GBM 

np.random.seed(0)
```

```python
# Example 1: Basic usage with synthetic data
# ------------------------------------------

# Initialize the GBM model
gbm = GBM()

# Set custom parameters for drift and volatility (with correlation)
drift = np.array([0.05, 0.03])
volatility = np.array([[0.1, 0.05], [0.05, 0.2]])
gbm.set_params(drift, volatility)

# Define initial values for the assets
initial_values = np.array([100, 150])

# Simulate paths
gbm.simulate_paths(S0=initial_values, N=200, T=2, n_sym=10)

# Plot the simulated paths
gbm.plot_paths(n_plots=2)
```

```python
# Example 2: Using real data (assumed here as numpy array)
# --------------------------------------------------------

# Suppose we have real observed data for two assets
# real_data = np.array([...])  # Replace with actual data

real_data = np.array([[100, 102, 101, 105, 110, 108, 115],
                 [103, 104, 105, 102, 108, 110, 109]])

# Initialize the GBM model
gbm_real = GBM()

# Fit the model to the real data
gbm_real.fit(real_data)

# Simulate paths based on the calibrated parameters
gbm_real.simulate_paths(N=200, T=2, n_sym=5)

# Plot the simulated paths
gbm_real.plot_paths(n_plots=6)

```
